```python
from tools import simulate, inital_state_delta, generate_matrix_simple
import matplotlib.pyplot as plt

# nicer looking default plots (according to your github)
plt.style.use('bmh')
```

```python
# Parameter
nodes = 21
starting_x = 10
connections = [-2,-1,1,2]
steps = 100
```

```python
# generate inital_state and matrix
intial_state = inital_state_delta(nodes,delta_x=starting_x)
t_matrix = generate_matrix_simple(nodes,connections)
```

```python
# simulate development
time_development = simulate(t_matrix,intial_state,steps)
```

```python
x = list(range(nodes))
plt.figure(figsize=(5,4))
for i in [5,10,20,50,100]:
    plt.scatter(x,time_development[i],label=f"{i}. step",alpha=0.7)

plt.xlabel("Node $i$")
plt.ylabel("Charge $V_i$")
plt.legend()
plt.savefig("images/task_b_time_dev.pdf")
```
