```python
from simulation import solve_simulation
from opt_initial_values import inital_values_gauss
from opt_diags_gen import diags_gen_backwards_euler, diags_gen_forward_euler, diags_gen_crank_nicholson
import matplotlib.pyplot as plt

# nicer looking default plots (according to your github)
plt.style.use('bmh')

import config as c
import plots as pl
```

```python
# forward euler
U_fe = solve_simulation(inital_values_gauss,diags_gen_forward_euler)
diff_times  = [0,4,50]
filename = "cable_euler_explicit"
pl.plot_diff_times(
    U_fe,
    t_diff_times=diff_times,
    log=False,
    alpha=0.8,
    legend=True,
    filename=filename
    )
```

```python
# backwards euler aka implicit
U_be = solve_simulation(inital_values_gauss,diags_gen_backwards_euler)

diff_times  = [1,5,50,100,199,c.n_t-1]
filename = "cable_euler_implicit"

pl.plot_diff_times(
    U_be,
    t_diff_times=diff_times,
    log=False,
    alpha=0.8,
    legend=True,
    filename=filename
    )
```

```python
# crank nicholson
U_cn = solve_simulation(inital_values_gauss,diags_gen_crank_nicholson)
diff_times  = [1,5,50,100,199,c.n_t-1]
filename = "cable_crank_nicholson"

pl.plot_diff_times(
    U_cn,
    t_diff_times=diff_times,
    log=False,
    alpha=0.8,
    legend=True,
    filename=filename
    )
```
